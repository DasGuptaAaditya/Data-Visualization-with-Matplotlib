Using Matplotlib for more EDA on Dataset

```python
import matplotlib. pyplot as plt

# Scatter plot with day against
plt.scatter (data['year'] ,data["age"])
# Adding Tit le to the Plot
plt.title ( "Scatter Plot" )
# Setting the X and Y labels
plt.xlabel(' Year')
plt.ylabel ("age")
plt.colorbar()
plt.show( )
```

```python
# Scatter plot with day against
plt.scatter (data['year'] ,data["income"])
# Adding Tit le to the Plot
plt.title ( "Scatter Plot" )
# Setting the X and Y labels
plt.xlabel(' Year')
plt.ylabel ("income")
plt.colorbar()
plt.show( )
```

```python
# Scatter plot with day against
plt.scatter (data['year'] ,data["size"])
# Adding Tit le to the Plot
plt.title ( "Scatter Plot" )
# Setting the X and Y labels
plt.xlabel(' Year')
plt.ylabel ("size")
plt.colorbar()
plt.show( )
```

```python

plt.bar (data['year'], data['expenditure'])
# Setting the X and Y labels
plt.xlabel ('year' )
plt.ylabel( 'expenditure')
# Adding the legends
plt.title ( "Bar Chart")
plt. show()
```

```python
# prompt: line chart

plt.plot(data['year'], data['expenditure'])
# Adding Title to the Plot
plt.title('Line Chart')
# Adding the X and Y labels
plt.xlabel('Year')
plt.ylabel('Expenditure')
# Showing the plot
plt.show()
```

```python
# prompt: heatmap

import seaborn as sns

plt.figure(figsize=(8, 5))
sns.heatmap(data.corr(), annot=True)
plt.show()
```

```python
# prompt: pie chart

plt.figure(figsize=(15, 10))
plt.pie(data['expenditure'], labels=data['year'])
plt.title('Expenditure by Year')
plt.show()
```
